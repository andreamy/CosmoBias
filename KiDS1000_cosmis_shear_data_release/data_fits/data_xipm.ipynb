```python
import astropy.io.fits as fits
import numpy as np
import random

from numpy.linalg import inv
from scipy.stats import chi2

cat_version = 'V1.0.0A_ugriZYJHKs_photoz_SG_mask_LF_svn_309c_2Dbins_v2_goldclasses_Flag_SOM_Fid'
```

```python
#############################################################################
# 2PCFs

stats_name ='xipm'

filename = stats_name+'_KIDS1000_BlindC_with_m_bias_'+cat_version+'.fits'
F=fits.open(filename)

# Covariance matrix includes covariances for xip, xim and their cross covariance
ext=F["COVMAT"]
covariance= ext.data

# cosmic shear data vector: xi_plus
ext=F['xip']
xipinfo = ext.data
xip=ext.data['VALUE']

# cosmic shear data vector: xi_minus
ext=F['xim']
xim=ext.data['VALUE']
theta = ext.data['ANG']

F.close()
```

```python
# Create a data vector with xip and xim, one after the other, ell and z ordering from low to high
datavector = np.hstack((xip,xim))
thetavector = np.hstack((theta,theta))
np.savetxt("data_xipm", np.transpose([thetavector, datavector]))
np.savetxt("theta_9bins_arcmin.txt", theta)
np.savetxt("kids_covariance_matrix", covariance)
```
