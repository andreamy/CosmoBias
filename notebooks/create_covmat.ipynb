This notebook loads from file the output covariance from CosmoCov, and converts it into a 2D array that can be later used by all the other Python codes/notebooks. It produces a color plot of the matrix as well

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import sys
```

#### Function to convert covariance matrix file into a numpy 2D array

```python
def get_cov(filename):
    
    """
    Convert covariance matrix file into a numpy 2D array.

    Parameters
    ----------
    filename : .txt file
                directory of the text file to be converted.

    Returns
    -------
    cov : 2darray
        Matrix containing the Gaussian contribution to the covariance.
        
    cov_ng : 2darray
        Matrix containing the Non- Gaussian contribution to the covariance. 
        
    ndata : int
        Dimension of the covariance matrix.
    """

    data = np.genfromtxt(filename)
    ndata = int(np.max(data[:,0]))+1

    print("Dimension of cov: %dx%d"%(ndata,ndata))

    ndata_min = int(np.min(data[:,0]))
    cov_g = np.zeros((ndata,ndata))
    cov_ng = np.zeros((ndata,ndata))
    
    for i in range(0,data.shape[0]):
        cov_g[int(data[i,0]),int(data[i,1])] =data[i,8]
        cov_g[int(data[i,1]),int(data[i,0])] =data[i,8]
        cov_ng[int(data[i,0]),int(data[i,1])] =data[i,9]
        cov_ng[int(data[i,1]),int(data[i,0])] =data[i,9]

    return cov_g, cov_ng, ndata
```

#### Choose simulation name and call get_cov(). Note that the covmat files are located at ../CosmoCov/covs/ for convenience. This can be changed but it has to be consistent within all files. 

```python
simulation = "class3.0_closer_derivative"
covfile = "../CosmoCov/covs/covmat_"+simulation
c_g, c_ng, ndata = get_cov(covfile)
cov = c_ng+c_g
np.save("CovMatrix_" + simulation, cov)
```

#### Check covariance is positive definite and plot correlation matrix

```python
b = np.sort(LA.eigvals(cov))
print("min+max eigenvalues cov: %e, %e"%(np.min(b), np.max(b)))
if(np.min(b)<=0.):
    print("non-positive eigenvalue encountered! Covariance Invalid!")
    exit()

print("Covariance is positive definite!")

cmap = 'seismic'

pp_norm = np.zeros((ndata,ndata))

for i in range(ndata):
    for j in range(ndata):
        pp_norm[i][j] = cov[i][j]/ np.sqrt(cov[i][i]*cov[j][j])

print("Plotting correlation matrix ...")

plot_path = '../Output/plots/covmat_'+simulation+'.pdf'
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(1, 1, 1)
#plt.axvline(x=20*20,color='black')
#plt.axvline(x=40*20,color='black')
#plt.axhline(y=20*20,color='black')
#plt.axhline(y=40*20,color='black')
im3 = ax.imshow(pp_norm, cmap=cmap, vmin=-1, vmax=1)
fig.colorbar(im3, orientation='vertical')
#ax.text(45, -15, r'$\xi_+^{ij}(\theta)$', fontsize=14)
#ax.text(175, -15, r'$\xi_-^{ij}(\theta)$', fontsize=14)
#
#ax.text(370, 75, r'$\xi_+$', fontsize=14)
#ax.text(370, 205, r'$\xi_-$', fontsize=14)
plt.savefig(plot_path,dpi=2000)
plt.show()
```
